```python
import pandas as pd
df=pd.read_csv('25totalcases.csv')
print(df)
```

```python
df=df.fillna(method='ffill')
df.shape
df.tail(1)
```

```python
df.dtypes
```

```python
df['date']=pd.to_datetime(df['date'])
df.dtypes
```

```python
df=df.set_index('date')
```

```python
df.head(3)
```

```python
grupo=df.groupby(df.index.month).tail(1).reset_index()
grupo
```

```python
grupo=grupo.T
```

```python
grupo
```

```python
grupo=grupo.drop(['date'])
grupo=grupo.reset_index()
grupo
```

```python
grupo.columns=['paises','diciembre','enero','febrero','marzo']
grupo
```

```python
import pycountry
def get_alpha_3(location):
 try:
   return pycountry.countries.get(name=location).alpha_3
 except:
   return None
grupo['code'] = grupo['paises'].apply(lambda x: get_alpha_3(x) )
print(grupo.head(5))
```

```python
dft=pd.read_csv('temperature.csv')
dft.head
```

```python
list(dft)
```

```python
dfgt = (grupo.merge(dft, left_on='code', right_on='ISO_3DIGIT')
         .reindex(columns=['paises', 'enero','febrero','marzo','code','Jan_Temp','Feb_temp','Mar_temp']))
print(dfgt.head(5))
```

```python
import plotly.express as px
fig = px.scatter(dfgt, x="Mar_temp", y="marzo", color="paises")
fig.show()
```

```python
import pandas as pd
import numpy as np
df=pd.read_csv('1Abriltotalcases.csv')
df.head()
```

```python
df=df.fillna(method='ffill')
primer=df.loc[(df['Mexico']>=1).idxmax(),'date']
primer
```

```python
df['date']=pd.to_datetime(df['date'])
cond=df.date>=pd.Timestamp(primer)
cond
```

```python
casos=df['Mexico'].loc[cond,]
casos
```

```python
len(casos)
```

```python
import numpy as np
fecha=np.arange(1,34,1)
len(fecha)
```

```python
import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(go.Scatter(y=casos, x=fecha,mode='lines',name='MX'))
fig.show()
```

```python
from pylab import *
from scipy.optimize import curve_fit
x=fecha
y=casos
def func(x, c, k):
   return (c*np.exp(k*x))
 
popt, pcov = curve_fit(func, x, y)
print(popt)
np.sqrt(np.diag(pcov))
```

```python
func(x,*popt)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(y=func(x,*popt), x=fecha,mode='lines',name='Ajuste'))
fig.add_trace(go.Scatter(y=casos, x=fecha,mode='lines',name='MX'))
fig.show()
```

```python
residuales=y-func(x,*popt)
residuales
```

```python
ss_res=np.sum(residuales**2)
ss_tot=np.sum(y-np.mean(func(x,*popt))**2)
r_sq=1-(ss_res/ss_tot)
print(r_sq)
```

```python
import pandas as pd
import numpy as np
df=pd.read_csv('5AbrilTotalCases.csv')
df=df.fillna(method='ffill')
df['date']=pd.to_datetime(df['date'])
df=df.set_index('date')
```

```python
resumen=pd.DataFrame()
resumen['PS']=df.Mexico.resample('W-MON').mean().round(2)
resumen
```

```python
resumen['PSm']=df.Mexico.resample('W-MON').min()
resumen
```

```python
resumen['S']=df.Mexico.resample('W-MON').last()
resumen
```

```python
df=df.filter(items=['Mexico', 'Spain','Italy','Chile','United States','Switzerland','Argentina','Colombia'])
pd.set_option('display.max_columns', None) 
pd.set_option('display.expand_frame_repr', False)
resumen2 = pd.DataFrame()
resumen2 = df.resample('M').last()
resumen2
```

```python
diferencias=df.diff()
diferencias=diferencias.abs()
list(diferencias)
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
cols_plot = ['Mexico', 'Spain','Italy','Chile','United States','Switzerland','Argentina','Colombia']
fig = make_subplots(rows=len(cols_plot), cols=1, start_cell="bottom-left")
i=1
j=0
for col in cols_plot:
 trace=go.Scatter(x=diferencias.index, y=diferencias[col],mode='lines',name=cols_plot[j])
 fig.append_trace(trace,i,1)
 i=i+1
 j=j+1
fig.show()
```
